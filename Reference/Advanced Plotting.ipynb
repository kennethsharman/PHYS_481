### Contour plots

```python
from matplotlib import pyplot as plt
import numpy as np

x_vals = np.linspace(-3, 3, 21)
y_vals = np.linspace(0, 10, 11)
X, Y = np.meshgrid(x_vals, y_vals)

# z(x, y) = cos(x)*sin(y)
Z = np.cos(X) * np.sin(Y)

cs = plt.contour(X, Y, Z, 10, linewidths=3, colors='k')
```

### Surface Plots

```python
from mpl_toolkits.mplot3d import Axes3D

ax = Axes3D( plt.figure() ) #

# If you have a small array then we need to use all points in it
# Python cheats when plotting so we may need the last two arguments
ax.plot_surface(X, Y, Z, rstride=1, cstride=1)
plt.show()
```

```python
points = np.linspace(-1, 1, 101)
X, Y = np.meshgrid(points, points)
Z = X**2 + Y**2
ax = Axes3D(plt.figure())
ax.plot_surface(X, Y, Z)
plt.show()
```
