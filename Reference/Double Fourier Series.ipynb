## Double Fourier Series

Let $F(x,y)$ be a function that is $2L$-periodic $(L>0)$ in $x$, and $2K$-periodic $(K>0)$ in $y$. This means that for any $x$, $y$ we have:

$$ F(x+2L, y) = F(x,y) \qquad and \qquad F(x,y+2k) = F(x,y) \qquad [1] $$

For every y-fixed, the Fourier series of $F(x,y)$ is 

$$ F(x,y) \approx \dfrac{1}{2} a_0(y) + \sum_{n=1}^{\infty} \Big[a_n(y) \cos{\Big( n \frac{\pi}{L} x \Big) } + b_n(y) \sin{\Big( n \frac{\pi}{L} x \Big)} \Big] $$ 

where the Fourier coefficients $a_n(y)$ and $b_n(y)$ are defined by

$$ a_n(y) = \frac{1}{L} \int_{0}^{2L} F(x,y) \cos{\Big( n \frac{\pi}{L} x \Big)} \; dx \qquad n = 0, 1, 2, \dots  $$

$$ b_n(y) = \frac{1}{L} \int_{0}^{2L} F(x,y) \sin{\Big( n \frac{\pi}{L} x \Big)} \; dx \qquad n = 1, 2, 3, \dots $$

From,

$$ a_n(y+2k) = \frac{1}{L} \int_{0}^{2L} F(x,y+2k) \cos{\Big( n \frac{\pi}{L} x \Big)} \; dx = \frac{1}{L} \int_{0}^{2L} F(x,y) \cos{\Big( n \frac{\pi}{L} x \Big)} \; dx = a_n(y) $$

$$ b_n(y+2k) = \frac{1}{L} \int_{0}^{2L} F(x,y+2k) \sin{\Big( n \frac{\pi}{L} x \Big)} \; dx = \frac{1}{L} \int_{0}^{2L} F(x,y) \sin{\Big( n \frac{\pi}{L} x \Big)} \; dx = b_n(y) $$

it follows that both $a_n(y)$ and $b_n(y)$ are $2k$-periodic. Their Fourier series are

$$ a(y) \approx \frac{1}{2} \alpha_{n0} + \sum_{m=1}^{\infty} \Big[ \alpha_{nm} \cos{\Big( m \frac{\pi}{k} y \Big)} + \gamma_{nm} \sin{\Big( m \frac{\pi}{k} y \Big)} \Big] \qquad [2] $$

$$ b(y) \approx \frac{1}{2} \beta_{n0} + \sum_{m=1}^{\infty} \Big[ \beta_{nm} \cos{\Big( m \frac{\pi}{k} y \Big)} + \delta_{nm} \sin{\Big( m \frac{\pi}{k} y \Big)} \Big] \qquad [3] $$ 

where,

$$ \alpha_{nm} = \frac{1}{k} \int_{0}^{2k} a_n(y) \cos{\Big( m \frac{\pi}{k} y \Big)} \; dy = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \cos \Big( n \frac{\pi}{L} x \Big) \cos \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 0, m \geq 0 $$

$$ \gamma_{nm} = \frac{1}{k} \int_{0}^{2k} a_n(y) \sin{\Big( m \frac{\pi}{k} y \Big)} \; dy = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \cos \Big( n \frac{\pi}{L} x \Big) \sin \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 0, m \geq 1 $$

$$ \beta_{nm} = \frac{1}{k} \int_{0}^{2k} b_n(y) \cos{\Big( m \frac{\pi}{k} y \Big)} \; dy = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \sin \Big( n \frac{\pi}{L} x \Big) \cos \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 1, m \geq 0 $$

$$ \delta_{nm} = \frac{1}{k} \int_{0}^{2k} b_n(y) \sin{\Big( m \frac{\pi}{k} y \Big)} \; dy = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \sin \Big( n \frac{\pi}{L} x \Big) \sin \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 1, m \geq 1 $$

Substituting [2] and [3] into [1] leads to

$$ \star \quad F(x,y)\approx 
\frac{1}{4} \alpha_{00} 
+ \frac{1}{2} \sum_{m=1}^{\infty} \Big[ \alpha_{0m} \cos \Big( m \frac{\pi}{k} y \Big) + \gamma_{0m} \sin \Big( m \frac{\pi}{k} y \Big) \Big] \\ 
+ \frac{1}{2} \sum_{n=1}^{\infty} \Big[ \alpha_{n0} \cos \Big( n \frac{\pi}{L} x \Big) + \beta_{n0} \sin \Big( n \frac{\pi}{L} x \Big) \Big]  \\
+ \sum_{n=1}^{\infty} \bigg[ \sum_{m=1}^{\infty} \alpha_{nm} \cos \Big( m \frac{\pi}{k} y \Big) + \gamma_{nm} \sin \Big( m \frac{\pi}{k} y \Big) \bigg] \cos \Big( n \frac{\pi}{L} x \Big) \\
+ \sum_{n=1}^{\infty} \bigg[ \sum_{m=1}^{\infty} \beta_{nm} \cos \Big( m \frac{\pi}{k} y \Big) + \delta_{nm} \sin \Big( m \frac{\pi}{k} y \Big) \bigg] \sin \Big( n \frac{\pi}{L} x \Big) $$

where,

$$ \alpha_{nm} = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \cos \Big( n \frac{\pi}{L} x \Big) \cos \Big( m \frac{\pi}{k} y\Big) \; dx \, dy \qquad n \geq 0, m \geq 0 $$

$$ \gamma_{nm} = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \cos \Big( n \frac{\pi}{L} x \Big) \sin \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 0, m \geq 1 $$

$$ \beta_{nm} = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \sin \Big( n \frac{\pi}{L} x \Big) \cos \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 1, m \geq 0 $$

$$ \delta_{nm} = \frac{1}{kL} \int_{0}^{2k} \int_{0}^{2L} F(x,y) \sin \Big( n \frac{\pi}{L} x \Big) \sin \Big( m \frac{\pi}{k} y \Big) \; dx \, dy \qquad n \geq 1, m\geq 1 $$

In order to evaluate the Double Fourier Series of a function, we will need to perform double integration. First, we define a double integral function.

```python
import scipy.integrate as integrate

def doubleIntegral(integrand, outerLo, outerHi, innerLo, innerHi, tol = 1e-6, divmax = 1000, isShow = False):
    '''
    Computes the value of a double integral using scipy's romberg
    Parameters:
        integrand: function that is integrated 
        outerLo, outerHi: lower and upper bounds for outer integral
        innerLo, innerHi: lower and upper bounds for inner integral
        tol, divmax: Integration and divmax tolerances for x and y integrations
        isShow: Show the triangular integration matrix for x and y integrals
    Returns value of the double integral
    '''
    
    def inner(x):
        return integrate.romberg(integrand, innerLo, innerHi, args=(x,), tol = tol, divmax = divmax, show = isShow)
    return integrate.romberg(inner, outerLo, outerHi, tol = tol, divmax = divmax, show = isShow)
```

$$ F(x,y)=xy \qquad 0<x<2\pi \quad 0<y<2\pi $$

It can be shown that

$$ F(x,y) \approx \frac{1}{4} \pi^2 - \frac{\pi}{2} \sum_{m=1}^{\infty} \frac{1}{m} \sin (my) - \frac{\pi}{2} \sum_{n=1}^{\infty} \frac{1}{n} \sin (nx) + \sum_{n=1}^{\infty} \sum_{m=1}^{\infty} \frac{1}{mn} \sin(my) \sin(nx) $$

I will define functions necessary to compute each term of this function.

```python
def sumsiney(y, m):
    result = 0
    while m != 0:
        result += (1.0/m) * np.sin(m*y)
        m -= 1
    return result

def sumsinex(x, n):
    result = 0
    while n != 0:
        result += (1.0/n) * np.sin(n*x)
        n -= 1
    return result

def summn(x, y, n, m):
    result = 0
    while n != 0:
        while m != 0:
            result += 1.0/(m*n) * np.sin(m*y) * np.sin(n*x)
            m -= 1
        n -= 1
    return result

def F(x,y, n, m):
    return 0.25 * np.pi**2 - (np.pi/2) * sumsiney(y, m) - (np.pi/2) * sumsinex(x, n) + summn(x, y, n, m)

def xfunc(x, n):
    return 0.25 * np.pi**2 - (np.pi/2) * sumsinex(x, n)

def actualF(x,y):
    return x*y
```
