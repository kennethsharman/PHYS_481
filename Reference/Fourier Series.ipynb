```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import quad

def Fourier(f, x, L, N):
    '''
    Calculates the Fourier Series of a function.
    Parameters:
        f - function to be approximated
        x - independent variable
        L - peroid of f
        N - number of iterations of reimann sum
    Returns:
        Fourier Series representation of f
    '''
    
    def fcos(x, f, L, n):
        '''Helper function. Defines integrand for a_n coefficient.'''
        return f(x) * np.cos( (n*np.pi/L)*x )

    def fsin(x, f, L, n):
        '''Helper function. Defines integrand for b_n coefficient.'''
        return f(x) * np.sin( (n*np.pi/L)*x )

    def a_0(f, L):
        '''Helper function. Computes the A_0 coefficient.'''
        return (1.0/(2.0*L)) * quad(f, -L, L, limit=1000)[0]

    def a_n(f, L, n):
        '''Helper function. Computes the A_n coefficient.'''
        return (1.0/L) * quad(fcos, -L, L, args=(f, L, n), limit=1000)[0]

    def b_n(f, L, n):
        '''Helper function. Computes the B_n coefficient.'''
        return (1.0/L) * quad(fsin, -L, L, args=(f, L, n), limit = 1000)[0]

    def F(f, x, L, n):
        '''Helper function. Computes the nth term of the Fourier Series representation of f.'''
        An = a_n(f, L, n) 
        Bn = b_n(f, L, n)
        return An * np.cos( (n*x*np.pi)/L ) + Bn * np.sin( (n*x*np.pi)/L )
    
    fourier_series = 0 # Initialize return variable to zero
    n_vals = np.linspace(1, N, N) # Create array of iteration numbers
    
    for n in n_vals:
        fourier_series += F(f, x, L, n) # Calcuate each term of series and add to return variable
    
    fourier_series += a_0(f, L) # Add A_0 coefficient to series
    
    return fourier_series
```

```python
def f(x):
    '''
    Define the function for which we will calculate its Fourier Series representation. Here, f(x) = x^2
    '''
    return x**2
```

```python
fig = plt.figure(figsize=(15,6))
ax = plt.subplot()

x_vals = np.linspace(-1.0, 1.0, 10**3)
plt.plot(x_vals, f(x_vals), label='$f(x)=x^2$', linewidth=6)
plt.plot(x_vals, Fourier(f, x_vals, 1.0, 100), label='Fourier Series', linestyle="--", color='red', linewidth=6)
ax.grid(linestyle='dashed')

plt.xlabel('x values', size='large')
plt.ylabel('y values', size='large')
plt.title('Fourier Series of f(x) = $x^2$', size='xx-large')

legend = plt.legend(loc='best', fontsize='x-large', shadow=(True), borderaxespad=3)
frame = legend.get_frame()
frame.set_facecolor('#f9f9f9')
frame.set_alpha(0.6)

plt.show()
```

```python
def g(x):
    '''
    Define the function for which we will calculate its Fourier Series representation. Here, f(x) = x^3
    '''
    return x**3
```

```python
def h(x):
    '''
    Define the function for which we will calculate its Fourier Series representation. Here, f(x) = x^2
    '''
    return x**3 - x * np.exp(x)
```

```python
fig = plt.figure(figsize=(15,10))
ax = plt.subplot()

x_vals = np.linspace(-5.0, 5.0, 10**3)
plt.plot(x_vals, h(x_vals), label='$h(x)=x^3-x*e^x$', linewidth=4)
plt.plot(x_vals, Fourier(h, x_vals, 5.0, 10), label='Fourier Series $n=10$', color='red', linewidth=2)
plt.plot(x_vals, Fourier(h, x_vals, 5.0, 100), label='Fourier Series $n=100$', color='purple', linewidth=2)
plt.plot(x_vals, Fourier(h, x_vals, 5.0, 500), label='Fourier Series $n=500$', linestyle='--', color='black', linewidth=2)
ax.grid(linestyle='dashed')

plt.xlabel('x values', size='large')
plt.ylabel('y values', size='large')
plt.title('Fourier Series of h(x) = $x^3-x*e^x$', size='xx-large')

legend = plt.legend(loc='best', fontsize='x-large', shadow=(True), borderaxespad=3)
frame = legend.get_frame()
frame.set_facecolor('#f9f9f9')
frame.set_alpha(0.6)

#plt.savefig('Fourier Series of x^3 - x * e^x')
plt.show()
```

```python
L = 1.0

def H(x):
    '''
    Heavyside function
    '''
    if x<0: return 0
    elif x==0: return 0.5
    else: return 1

def sqrwave(x):
    '''
    Ideal square wave function, with period L.
    '''
    return 2.0 * ( H(x/L) - H(x/L - 1) ) -1.0
```

```python
fig = plt.figure(figsize=(15,7))
ax = plt.subplot()

x_vals = np.linspace(-1.0, 1.0, 10**3)
plt.plot(x_vals, Fourier(sqrwave, x_vals, L, 200), label='Fourier Series n = 200', color='green', linewidth=2)
plt.plot(x_vals, Fourier(sqrwave, x_vals, L, 10), label='Fourier Series n = 10', color='yellow', linewidth=3)
plt.plot(x_vals, Fourier(sqrwave, x_vals, L, 5), label='Fourier Series n = 5', color='blue', linewidth=3)
plt.plot(x_vals, Fourier(sqrwave, x_vals, L, 3), label='Fourier Series n = 3', color='magenta', linewidth=3)
plt.plot(x_vals, Fourier(sqrwave, x_vals, L, 1), label='Fourier Series n = 1', color='red', linewidth=3)

ax.grid(linestyle='dashed')

plt.xlabel('x values', size='large')
plt.ylabel('y values', size='large')
plt.title('Fourier Series of a Square Wave', size='xx-large')

legend = plt.legend(loc='best', fontsize='x-large', shadow=(True), borderaxespad=3)
frame = legend.get_frame()
frame.set_facecolor('#f9f9f9')
frame.set_alpha(0.6)

plt.show()
```
